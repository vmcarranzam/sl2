```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Embedding, LSTM, Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

```python
train_data = pd.read_csv('twitter_training.csv')
validation_data = pd.read_csv('twitter_validation.csv')
```

```python
train_data = train_data.rename(columns={
    '2401': 'number',
    'Borderlands': 'borderline',
    'Positive': 'sentiment',
    'im getting on borderlands and i will murder you all ,': 'text'
})

validation_data = validation_data.rename(columns={
    '3364': 'number',
    'Facebook': 'borderline',
    'Irrelevant': 'sentiment',
    "I mentioned on Facebook that I was struggling for motivation to go for a run the other day, which has been translated by Tomâ€™s great auntie as â€˜Hayley canâ€™t get out of bedâ€™ and told to his grandma, who now thinks Iâ€™m a lazy, terrible person ðŸ¤£": 'text'})

```

```python
train_data = train_data.drop(["number","borderline"], axis = 1)
```

```python
validation_data = validation_data.drop(["number","borderline"], axis = 1)
```

```python
train_data.head()
```

```python
validation_data.head()
```

```python
text_data = train_data['text']
labels = train_data['sentiment']
validation_text_data = validation_data['text']
validation_labels = validation_data['sentiment']
```

```python
label_encoder = LabelEncoder()
labels_encoded = label_encoder.fit_transform(labels)
validation_labels_encoded = label_encoder.transform(validation_labels)
```

```python
model = keras.Sequential()
model.add(Embedding(input_dim=10, output_dim=10, input_length=10))
model.add(LSTM(units=128))
model.add(Dense(units=10, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(text_data, labels_encoded, epochs=100, batch_size=10, validation_data=(validation_text_data, validation_labels_encoded))
```

```python
# Evaluate the model
test_loss, test_accuracy = model.evaluate(validation_text_data, validation_labels_encoded)
print(f'Test accuracy: {test_accuracy}')
```
